# Running the Statistics - Air Temp

```python
el['Air Temp'] = pd.to_numeric(el['Air Temp'], errors='coerce')
el['Air Temp'].describe()
```

# Running the Statistics - Sea Temp

```python
el['Sea Surface Temp'] = pd.to_numeric(el['Sea Surface Temp'], errors='coerce')
el['Sea Surface Temp'].describe()
```

# Performing the Plot

```python
sns.jointplot(x = 'Air Temp', y = 'Sea Surface Temp', data = el, color='green', size = 10, alpha = 0.4)
```
